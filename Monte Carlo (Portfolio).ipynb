```python
import pandas_datareader as pdr
import datetime as dt
import pandas as pd
import numpy as np
```

```python
tickers = ['ROSN','NLMK','MTLRP','CHMF','MOEX','SBER','PHOR','MAGN','YNDX','TATNP','TCSG','SNGS','TATN']
start = dt.datetime(2020, 1, 1)
cols=['CLOSE']
data = pd.DataFrame()
for com in tickers:
    df = pdr.get_data_moex(com,start)
    df = df[['CLOSE']]
    df = df.rename(columns={'CLOSE': com})
    data = pd.concat([data, df], axis=1)
```

```python
data
```

```python
count=len(tickers)
```

```python
log_returns = np.log(data/data.shift())
```

```python
log_returns
```

```python
weight = np.random.random(count)
weight /= weight.sum()
weight
```

```python
exp_rtn = np.sum(log_returns.mean()*weight)*252
```

```python
exp_vol = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252, weight)))
sharpe_ratio = exp_rtn / exp_vol
sharpe_ratio
```

```python
n = 10000

weights = np.zeros((n, count))
exp_rtns = np.zeros(n)
exp_vols = np.zeros(n)
sharpe_ratios = np.zeros(n)

for i in range(n):
    weight = np.random.random(count)
    weight /= weight.sum()
    weights[i] = weight
    
    exp_rtns[i] = np.sum(log_returns.mean()*weight)*252
    exp_vols[i] = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252, weight)))
    sharpe_ratios[i] = exp_rtns[i] / exp_vols[i]
```

```python
sharpe_ratios.max()
```

```python
sharpe_ratios.argmax()
```

```python
weights[sharpe_ratios.argmax()]
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
fig, ax = plt.subplots()
ax.scatter(exp_vols, exp_rtns, c=sharpe_ratios)
ax.scatter(exp_vols[sharpe_ratios.argmax()], exp_rtns[sharpe_ratios.argmax()], c='r')
ax.set_xlabel('Expected Volatility')
ax.set_ylabel('Expected Return')
```
